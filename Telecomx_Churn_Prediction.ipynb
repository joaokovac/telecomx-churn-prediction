```python
# ==========================================
# TELECOM X - PREVISÃO DE CHURN
# DESAFIO DATA SCIENCE ALURA
# ==========================================

%matplotlib inline

# IMPORTAÇÃO DAS BIBLIOTECAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


# ==========================================
# CARREGAR DATASET
# ==========================================


print("Primeiras linhas do dataset:")
print(df.head())

print("\nInformações do dataset:")
print(df.info())

print("\nEstatísticas gerais:")
print(df.describe())


# ==========================================
# VERIFICAR VALORES NULOS
# ==========================================

print("\nValores nulos por coluna:")
print(df.isnull().sum())


# ==========================================
# DISTRIBUIÇÃO DE CHURN
# ==========================================

plt.figure()
sns.countplot(x="Churn", data=df)
plt.title("Distribuição de Churn")
plt.xlabel("Churn")
plt.ylabel("Quantidade de clientes")
plt.show()


# ==========================================
# HISTOGRAMA DAS VARIÁVEIS NUMÉRICAS
# ==========================================

df.select_dtypes(include=['int64','float64']).hist(figsize=(15,10))
plt.suptitle("Distribuição das variáveis numéricas")
plt.show()


# ==========================================
# BOXPLOTS DAS VARIÁVEIS NUMÉRICAS
# ==========================================

for coluna in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure()
    sns.boxplot(x=df[coluna])
    plt.title(f"Boxplot da variável {coluna}")
    plt.show()


# ==========================================
# CORRELAÇÃO ENTRE VARIÁVEIS NUMÉRICAS
# ==========================================

df_corr = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(12,8))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de Correlação entre variáveis")
plt.show()


# ==========================================
# CONVERTER VARIÁVEIS CATEGÓRICAS
# ==========================================

df_modelo = pd.get_dummies(df, drop_first=True)


# ==========================================
# SEPARAÇÃO DAS VARIÁVEIS
# ==========================================

X = df_modelo.drop("Churn", axis=1)
y = df_modelo["Churn"]


# ==========================================
# DIVISÃO TREINO E TESTE
# ==========================================

X_treino, X_teste, y_treino, y_teste = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)


# ==========================================
# NORMALIZAÇÃO
# ==========================================

scaler = StandardScaler()

X_treino = scaler.fit_transform(X_treino)
X_teste = scaler.transform(X_teste)


# ==========================================
# MODELO DE MACHINE LEARNING
# ==========================================

modelo = LogisticRegression(max_iter=1000)

modelo.fit(X_treino, y_treino)


# ==========================================
# PREVISÕES
# ==========================================

previsoes = modelo.predict(X_teste)


# ==========================================
# AVALIAÇÃO DO MODELO
# ==========================================

acuracia = accuracy_score(y_teste, previsoes)

print("\nAcurácia do modelo:", acuracia)

print("\nRelatório de classificação:")
print(classification_report(y_teste, previsoes))


# ==========================================
# MATRIZ DE CONFUSÃO
# ==========================================

matriz = confusion_matrix(y_teste, previsoes)

plt.figure()
sns.heatmap(matriz, annot=True, fmt="d")
plt.title("Matriz de Confusão")
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.show()


# ==========================================
# IMPORTÂNCIA DAS VARIÁVEIS
# ==========================================

coeficientes = pd.Series(modelo.coef_[0], index=X.columns)

coeficientes = coeficientes.sort_values()

plt.figure(figsize=(10,6))
coeficientes.plot(kind="barh")
plt.title("Importância das variáveis no modelo")
plt.xlabel("Impacto no churn")
plt.show()


# ==========================================
# CONCLUSÃO
# ==========================================

print("\nConclusão da análise:")
print("O modelo de regressão logística foi utilizado para prever o churn de clientes.")
print("A análise exploratória permitiu identificar padrões entre as variáveis.")
print("Com base no modelo, é possível prever clientes com maior risco de cancelamento.")
print("Essas informações podem ajudar a empresa a criar estratégias de retenção.")
```
